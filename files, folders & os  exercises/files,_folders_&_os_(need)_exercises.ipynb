```python
import glob
import os
import datetime

os_collaboration_exercise_data = '~/Desktop/NUS/NUS Main/Y2S2/SP2273/learning-portfolio-tchinqi/os-collaboration-exercise-data/'

os_collaboration_exercise_data = os.path.expanduser(os_collaboration_exercise_data)

months = {
    'jan': '01', 'feb': '02', 'mar': '03', 'apr': '04', 'may': '05', 
    'jun': '06', 'jul': '07', 'aug': '08', 'sep': '09', 'oct': '10', 
    'nov': '11', 'dec': '12'
}

def rename_my_file(old_file_name):
    parts = old_file_name.split('_')
    date_part, city_part_with_ext = parts[0], parts[1]
    city_part, ext = os.path.splitext(city_part_with_ext)  # Separates city and extension
    try:
        month_abbr, day = date_part.split('-')
        month_num = months[month_abbr.lower()]  # Converts month to number
        new_filename = f"{month_num}-{day}_{city_part}{ext}"  # Reassembles filename
    except KeyError:
        # If month abbreviation not found, return the original filename
        return old_file_name
    return new_filename

# matching all txt files in the folder
pattern = os.path.join(os_collaboration_exercise_data, '*.txt')

folder_path = os_collaboration_exercise_data

for old_path in glob.glob(pattern):
    old_file_name = os.path.basename(old_path)
    new_file_name = rename_my_file(old_file_name)
    if new_file_name and new_file_name != old_file_name:
        new_path = os.path.join(folder_path, new_file_name)
        os.rename(old_path, new_path)
        print(f"Renamed '{old_file_name}' to '{new_file_name}'")
```

```python
import os
import glob

os_collaboration_exercise_data = '~/Desktop/NUS/NUS Main/Y2S2/SP2273/learning-portfolio-tchinqi/os-collaboration-exercise-data/'
os_collaboration_exercise_data = os.path.expanduser(os_collaboration_exercise_data)

pattern = os.path.join(os_collaboration_exercise_data, '*.txt')

#store cities names
cities = set()

# extract city names from file
for file_path in glob.glob(pattern):
    filename = os.path.basename(file_path)
    parts = filename.split('_')
    if len(parts) > 1:
        city_with_extension = parts[1]
        city, _ = os.path.splitext(city_with_extension)
        cities.add(city)

# create a folder for each unique city
for city in cities:
    city_folder_path = os.path.join(os_collaboration_exercise_data, city)
    if not os.path.exists(city_folder_path):
        os.makedirs(city_folder_path)
        print(f"Created folder: {city_folder_path}")
    else:
        print(f"Folder exists: {city_folder_path}")
```

```python
#Attempt at one code to move all files into respective folders
import shutil
import os
import glob

source_directory = '~/Desktop/NUS/NUS Main/Y2S2/SP2273/learning-portfolio-tchinqi/os-collaboration-exercise-data/'
source_directory = os.path.expanduser(source_directory)

cities = ['cairo', 'dhaka', 'jakarta', 'karachi', 'manila', 'melbourne', 'osaka', 'shanghai', 'singapore', 'tokyo']

for city in cities:
    target_directory = os.path.join(source_directory, city)
    os.makedirs(target_directory, exist_ok=True)  # Create the directory if it doesn't exist
    
    pattern = os.path.join(source_directory, f'*{city}*.*')

    for file_path in glob.glob(pattern):
        filename = os.path.basename(file_path)
        target_path = os.path.join(target_directory, filename)
        shutil.move(file_path, target_path)
        print(f"Moved '{filename}' to '{target_path}'!")
```
