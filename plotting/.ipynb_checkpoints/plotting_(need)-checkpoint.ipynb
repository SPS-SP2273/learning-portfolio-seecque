# A simple plot

## Letâ€™s look at some code.

```python
#Here is some code and the resulting plot
from matplotlib import pyplot as plt
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, label='Y values',
         color='red', linestyle='dashed')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

### Things to note

1. We can use the following abbreviations if we like:

|Long form|Abbreviation|
|:-------:|:-----------|
|`color`|`c`|
|`linestyle`|`ls`|
|`linewidth`|`lw`|

```python
#As such both the lines produce the same result:
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
plt.plot(x, y, c='red', ls='dashed', lw=2)
```

2. Jupyter is an interactive environment, so we will see an output even if we omit `plt.show()`. However, it is good practice to include this line anyways so our code will also work in a non-interactive environment.
3. The plotting functions usually have default values for the styling parameter. So if we wish, we can keep it simple and plot just using:

```python
plt.plot(x, y_1, y_2)
```

4. We can split the arguments into separate lines to improve readability. So both the following forms are acceptable:

```python
plt.plot(x, y, color='red', linestyle='dashed', linewidth=2)
plt.plot(x, y_1, label='Y values',
           color='red', linestyle='dashed')
```

5. The order of how we specify the keyword arguments (`color`, `linewidth`,...) does not matter!

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed')
plt.plot(x, y_2, label='Y2 values', color='blue')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')
```

## 1.3 Yet another plot but with error bars

Here we add another plot, but we will include *x* and *y* error bars for the points. <br>The plotting command I need to use for this is called `errorbar()`. 

```python

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]
y_3 = [0, 4, 8, 12, 16, 20, 24]
x_error, y_error = .1, 0.75

# Lets start plotting
plt.plot(x, y_1, label='Y1 values', color='red', linestyle='dashed',)
plt.plot(x, y_2, label='Y2 values', color='blue', )
plt.errorbar(x, y_3, xerr=x_error, yerr=y_error,
             label='Y3 with errors', color='green')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc='upper left')

#code for error bar plotting:
plt.errorbar(x, y_3,
             xerr=x_error, yerr=y_error,
             label="Y3 with errors",
             color="green")
```

# 2 Better with NumPy

Often it is easier to use Numpy arrays than Python lists. So let's convert pyton lists to Numpy arrays then redo the plot in the previous step!

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

import numpy as np 
np_x = np.array(x)
np_y_1 = np.array(y_1)
np_y_2 = np.array(y_2)

plt.plot(np_x, np_y_1, color='red', linestyle='dashed', label='Y values')
plt.plot(np_x, np_y_2, color='blue', label='Y2 values')
```

## 2.1 Adding mathematical functions

One of the advantages of NumPy arrays is that they allow us to generate data-related mathematical functions easily. Let's reuse our previous code to plot $x^2$ and $sin(x)$

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])

x2 = x**2                  # The math stuff
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed', )
plt.plot(x, sin_x, label='sin(x)',
         color='blue')
plt.legend()                                 
```

Our plot doesn't look too good because sin(x) lies within $\pm1$, but $x^2$ has no such bounds. A way to fix this is to add another y-axis that shares the same x-axis.

## We need another axis!

Matplotlib offers a variety of ways to have multiple axes. The simplest way is to have another y-axis that shares the same x-axis. <br>
We can use the command `twinx()` for this

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                  # For y-axis 1

plt.twinx()                                                   # This creates a new y-axis 
                                                              # for the plots that comes after
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')                                 # For y-axis 2


#Qn for reviewer: How come my environment only gives 1 output?
```

### Things to note

- We now have two legend() calls, one for each axis. 
- Our plot still does not look good because we only have a few points.
  - Let's use `np.linspace` to fix this:

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',color='red', linestyle='dashed')
plt.legend(loc='lower left')                                

plt.twinx()                                                  
                                                            
plt.plot(x, sin_x, label='sin(x)',color='blue', )
plt.legend(loc='lower right')
```

# 3 Saving to disc

If we want to use our plot in a report or a presentation, we must first save it to a disk. Luckily, matplotlib allows us to export these plots into many formats (PDF, JPEG, PNG, BMP, etc) and at different resolutions too. 

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, label='x^2',
         color='red', linestyle='dashed')
plt.legend(loc='lower left')
plt.twinx()
plt.plot(x, sin_x,
         label='sin x', color='blue')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)
```

Here is how to export our plot into our disk:
1. We need to use the function `savefig()` for this
2. We specify the format with the extension (e.g. `filename.pdf`) of the file name,
3. We specfiy resolution by use dots-per-inch (`dpi`). 

When we run this code, we will find the file saved in the same directory as the one our notebook lives in. If we want it saved somewhere else, we should specify.
For example:<br>
`plt.savefig('~/Desktop/simple-plot.png', dpi=150)`

# 4 A real example: Global Warming

## 4.1 Plotting data from files

Plotting data stored from a file (e.g. spreadsheet, text file, database) is a routine task for a scientist. The first thing we should do with any data is to look at it with a simple plot.

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
```

```python
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
#Adding horizontal lines at the zero value to highlight the trend shown by the data.
#The 'hlines()' function needs a y-value and starting and ending values for x
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

## 4.2 Styles

To customise the plots, Matplotlib offers some standard style templates. (see [<u>here</u>](https://tonysyu.github.io/raw_content/matplotlib-style-gallery/gallery.html))

```python
plt.style.use('fivethirtyeight')
#Include the line at the top!
data = np.loadtxt('Complete_TAVG_daily.txt',
                   skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
#Adding horizontal lines at the zero value to highlight the trend shown by the data.
#The 'hlines()' function needs a y-value and starting and ending values for x
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```
