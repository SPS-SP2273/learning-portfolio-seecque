# Comparing the two ‘dialects’

To compare the 2 dialects, we will create the same plot using both `pyplot` and OO dialects, so that we can see how the latter works!

```python
#generate data for the plot:
import numpy as np
x = np.linspace(-np.pi, np.pi, num=100)
cos_x = np.cos(x)
sin_x = np.sin(x)
```

```python
#Comparison: 
#pyplot
from matplotlib import pyplot as plt
plt.plot(x, cos_x, label='cos x')
plt.plot(x, sin_x, label='sin x')
plt.legend()

#OO 
fig, ax = plt.subplots(nrows=1, ncols=1)
ax.plot(x, cos_x, label='cos x')
ax.plot(x, sin_x, label='sin x')
ax.legend()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=1)

ax[0].plot(x, cos_x, label='cos x')
ax[1].plot(x, sin_x, label='sin x')

ax[0].legend()
ax[1].legend()
```

# A complete OO example

The following is a simple example that creates a nicer, tweaked version of the previous plot.

```python
fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
fig.tight_layout()
```

```python
#Explanation of the above code:
fig, ax = plt.subplots(nrows=2, ncols=1, #Setting up the axes and figures
                       figsize=(5, 5), #Change the size of the figure by specifying a figure size
                       sharex=True) #Ask that the plots share the same axis by using 'sharex'
ax[0].plot(x, cos_x, label=r'$\cos(x)$') 
ax[0].fill_between(x, 0, cos_x, alpha=.25) #since x is a numpy array, we can index them using 0 and 1, and then we plot to that axis. 
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25) #Use fill_between to fill the plots

#draw the legends, we can do it one axis at a time.
# a more sensible way to do this is with a for loop that iterates through the item 'ax'
for a in ax: 
    a.legend()
    a.grid(alpha=.25) #we also add a grid to each plot and set the label 
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$') #since we are sharing the x axis, we only need to lavel the lowest plot, which has the index 1. 

fig.suptitle(r'$\sin(x)$ and $\cos(x)$') #set title, to figure not plot!
fig.tight_layout() #we use tight_layout() to ask Matplot to make any necessary adjustments to the layout to make out plot look nice
```

```python
#Without tight_layout():

fig, ax = plt.subplots(nrows=2, ncols=1,
                       figsize=(5, 5),
                       sharex=True)
ax[0].plot(x, cos_x, label=r'$\cos(x)$')
ax[0].fill_between(x, 0, cos_x, alpha=.25)
ax[1].plot(x, sin_x, label=r'$\sin(x)$')
ax[1].fill_between(x, 0, sin_x, alpha=.25)

for a in ax:
    a.legend()
    a.grid(alpha=.25)
    a.set_ylabel('$y$', rotation=0)

ax[1].set_xlabel('$x$')

fig.suptitle(r'$\sin(x)$ and $\cos(x)$')
```

<span style='color:darkred'>Note! *Pyplot* and *OO* interfaces do not use identical function names. For example, pyplot used *xlabel()* to set $x$ label but OO uses set_xlabel()! </span>

# Multiple rows and columns

Here we learn how to work with multiple rows and columns that forms a grid of plots. 

```python
fig, ax = plt.subplots(nrows=2, ncols=2,
                       figsize=(5, 5),
                       sharex='col', sharey='col')

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

As seen above, Matplotlib has autoscaled both axes.

Here are some way to specify how to share the axes:
<br>
|Option|Result|
|:----:|:-----|
|`True`|Makes **all** the axes use the same range|
|`col`|Use the same range for all the columns|
|`row`|Use the same range for all the rows|

```python
#Let's try the following:
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'
)

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

All the plots have the same range for the $x$ axis. 

Using `shareex='col'` is more suited for the data we're plotting. 
* It is important to know that how you decide on the most correct depends on what story you are trying to communicate with your plot.

```python
#Using sharex='col'
fig, ax = plt.subplots(
    nrows=2, ncols=2,
    figsize=(5, 5),
    sharex='col', sharey='row'
)

# Some variables to access the axes and improve readabilty
top_left, top_right, bottom_left, bottom_right = ax.flatten()

top_left.set_title('Top Left')
top_right.set_title('Top Right')
bottom_left.set_title('Bottom Left')
bottom_right.set_title('Bottom Right')

top_left.hlines(y=1, xmin=0, xmax=4)
top_right.hlines(y=2, xmin=0, xmax=5)
bottom_left.hlines(y=3, xmin=0, xmax=4)
bottom_right.hlines(y=4, xmin=0, xmax=5)

for a in ax.flatten():
    a.grid(alpha=.25)

plt.tight_layout()
```

## A quick helper function

We will need to read the data from these files several times, so we should create a function called `det_plot_data()` that we can call. 

```python
import numpy as np
def get_plot_data():
    data = {}
    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_all.csv'
    data['All'] = np.loadtxt(filename, skiprows=1)

    filename = 'sg-gov-graduate-employment-survey_basic_monthly_mean_nus.csv'
    data['NUS'] = np.loadtxt(filename, skiprows=1)

    return data
```

## The histogram

```python
plt.style.use('bmh')
data = get_plot_data()

# bins specifies how many bins to split the data
plt.hist([data['All'], data['NUS']], bins=50, label=['All', 'NUS'])
plt.xlabel('Mean of Basic Montly Earning (S$)')
plt.ylabel('Number of Students')
plt.legend()
```

## Scatter plots

Scatterplots are created by putting a marker at ($x$, $y$) point we specify. But since we need some values for $x$, we are going to use `range()` along with `len()` to generate a list `[0,1,2]`.... (appropriate to the dataset ofcourse)

```python
plt.style.use("seaborn-v0_8-darkgrid")

data = get_plot_data()

for label, numbers in data.items():
    x = range(len(numbers))
    y = numbers
    plt.scatter(x, y, label=label, alpha=.5)

plt.xlabel('Position in the list')
plt.ylabel('Mean of Basic Montly Eraning (S$)')
plt.legend()
```

## Pie charts

```python
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

plt.style.use('fivethirtyeight')
plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90                
        )
plt.title('Percentage of each major')
```

```python
#Changing colours of pie chart

import matplotlib.pyplot as plt

# Data
student_numbers = {'Life Sciences': 14,
                   'Physics': 12,
                   'Chemistry': 8,
                   'Comp. Biology': 1}
majors = list(student_numbers.keys())
numbers = list(student_numbers.values())

#Theme
plt.style.use('fivethirtyeight')

#Colors used
colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99']

plt.pie(numbers, 
        labels=majors,
        autopct='%1.1f%%',   # How to format the percentages
        startangle=-90,
        colors=colors       # Assigning custom colors
       )

plt.title('Percentage of each major')

plt.show()
```
